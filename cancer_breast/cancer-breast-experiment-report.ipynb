```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
#Load datasets

xgboost_df = pd.read_csv("./results/xgboost/cancer-breast-xgboost-report.csv")
lightgbm_df = pd.read_csv("./results/lightgbm/cancer-breast-lightgbm-report.csv")
catboost_df = pd.read_csv("./results/catboost/cancer-breast-catboost-report.csv")
random_forest_df = pd.read_csv("./results/random-forest/cancer-breast-random-forest-report.csv")
```

```python
def plot_accuracy_chart():
    xgboost = xgboost_df['Accuracy'].tolist() # list(map(lambda x: x * 100, xgboost_df['Accuracy'].tolist())
    lightgbm = lightgbm_df['Accuracy'].tolist()
    catboost = catboost_df['Accuracy'].tolist()
    random_forest = random_forest_df['Accuracy'].tolist()
    
    plt.plot(list(range(1,(len(xgboost) + 1))), xgboost, label = "XGBoost")
    plt.plot(list(range(1,(len(lightgbm) + 1))), lightgbm, label = "LightGBM")
    plt.plot(list(range(1,(len(catboost) + 1))), catboost, label = "Catboost")
    plt.plot(list(range(1,(len(random_forest) + 1))), random_forest, label = "Random Forest")
    
    plt.xlabel("Number of features")
    plt.ylabel("Accuracy (%)")  
    plt.legend()

    plt.savefig('./results/final-report/cancer-breast-accuracy.png')
             

plot_accuracy_chart()
```
